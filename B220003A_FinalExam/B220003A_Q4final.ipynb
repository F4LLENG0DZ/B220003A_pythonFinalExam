```python
import pandas as pd
```

```python
df = pd.read_csv('happy.csv')
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

gdp_sorted_df = df.sort_values(by = 'Log GDP per capita', ascending=True)
gen_sorted_df = df.sort_values(by = 'Generosity', ascending=True)

x = gdp_sorted_df['Log GDP per capita'].query("'Regional indicator' == Western Europe")
y = gen_sorted_df['Generosity'].query("'Regional indicator' == Western Europe")

fig, ax = plt.subplots()
ax.plot(x,y)
```
